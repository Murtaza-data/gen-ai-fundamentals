```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Some sample sentences
sentences = [
    "The dog is running in the park",
    "A puppy is jogging outside",
    "I love eating pizza"
]

# Convert text to embeddings (vectors)
embeddings = model.encode(sentences)

print("Shape:", embeddings.shape)
print("\nFirst embedding (first 10 numbers):", embeddings[0][:10])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compare all sentences against each other
similarity = cosine_similarity(embeddings)

print("Similarity between 'dog running' and 'puppy jogging':", round(similarity[0][1], 4))
print("Similarity between 'dog running' and 'pizza':", round(similarity[0][2], 4))
```

```python
import chromadb

# Create a ChromaDB client (runs locally, no API key needed)
client = chromadb.Client()

# Create a collection (like a table in a regular database)
collection = client.create_collection("my_first_vectordb")

# Some sample documents
documents = [
    "Python is a popular programming language for data science",
    "Machine learning models learn patterns from data",
    "ChromaDB is a vector database for storing embeddings",
    "Football is a popular sport played worldwide",
    "Deep learning uses neural networks with many layers"
]

# Add documents to the collection
collection.add(
    documents=documents,
    ids=["doc1", "doc2", "doc3", "doc4", "doc5"]
)

print("Documents stored successfully!")
print("Total documents in DB:", collection.count())
```

```python
# Ask a question
results = collection.query(
    query_texts=["What database stores embeddings?"],
    n_results=2  # return top 2 most similar documents
)

print("Top matching documents:")
for i, doc in enumerate(results['documents'][0]):
    print(f"\nResult {i+1}: {doc}")
```
